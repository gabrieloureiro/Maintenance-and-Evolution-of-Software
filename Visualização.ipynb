```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (20,20)
```

# APACHE

```python
apache = pd.read_csv('apache_cc.csv')
```

```python
columns = []
for col in apache.columns:
    if("|__coupled" in col):
        columns.append(col)
```

```python
for col in columns:
    apache.drop(col, inplace=True, axis=1)
```

```python
apache.describe()
```

```python
plt.hist(np.array(apache['|__loc']), np.arange(np.min(np.array(apache['|__loc'])), np.max(np.array(apache['|__loc'])), 50)) 
plt.title("LOC") 
plt.show()
```

```python
plt.hist(np.array(apache['|__fanin']), np.arange(np.min(np.array(apache['|__fanin'])), np.max(np.array(apache['|__fanin'])),2.5)) 
plt.title("FANIN") 
plt.show()
```

```python
plt.hist(np.array(apache['|__fanout']), np.arange(np.min(np.array(apache['|__fanout'])), np.max(np.array(apache['|__fanout'])),2)) 
plt.title("FANOUT") 
plt.show()
```
